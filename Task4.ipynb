We can see there are too much number of missing entries in Unnamed:2,Unnamed:3,Unnamed:4 columns i.e more than 99%. So we should have to remove these column.

```python
df4.drop(columns=df4[['Unnamed: 2','Unnamed: 3','Unnamed: 4']],axis=1,inplace=True)
```

```python
df4
```

```python
df4.shape
```

```python
df4.columns = ['spam/ham' , 'descrip']
```

```python
df4['spam/ham'] = df4['spam/ham'].replace({'spam': 0, 'ham': 1})
#convert text to numerical
```

```python
df4
```

```python
a = df4.descrip
```

```python
a
```

```python
b = df4['spam/ham']
b
```

```python
from sklearn.model_selection import train_test_split
```

```python
atrain,atest,btrain,btest=train_test_split(a,b,test_size=0.2,random_state=3)
```

```python
print(a.shape)
print(atrain.shape)
print(atest.shape)
```

```python
atrain,atest
```

```python
btrain,btest
```

Machine learning algorithms only performs well with respect to numbers,so we need to convert all the text data into numbers.To do so I will use TfidfVectorizer techinque from feature_extraction of sklearn.

```python
feat=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
feat
```

```python
btrain=btrain.astype('int')
btest=btest.astype('int')
```

```python
atrain_vec =feat.fit_transform(atrain)
```

```python
atest_vec =feat.transform(atest)
```

```python
print(atrain)
```

```python
atrain_vec
```

```python
print(atrain_vec)
```

```python
print(atest_vec)
```

```python
Log=LogisticRegression()
```

```python
Log.fit(atrain_vec,btrain)
```

```python
Log.score(atrain_vec,btrain)
```

```python
Log.score(atest_vec,btest)
```

```python
pre_Log=Log.predict(atest_vec)
pre_Log
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
accuracy_score(btest,pre_Log)
```

```python
confusion_matrix(btest,pre_Log)
```

```python
print(classification_report(btest,pre_Log))
```
